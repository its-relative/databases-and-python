```python
import pandas as pd
import mysql.connector
from sqlalchemy import create_engine

# Step 1: Read CSV file into a pandas DataFrame
csv_file_path = '/home/its_relative/folder/in_git_hub/databases-and-python/mysql/excelr/assignments/1st/orders.csv'
df = pd.read_csv(csv_file_path, encoding='latin1')

# Step 2: Create connection to MySQL database
cnx = mysql.connector.connect(
    user='root',
    password='sapy@1234',
    host='localhost',
    database='assgnments'
)
cursor = cnx.cursor()

# Step 3: Function to map pandas dtypes to MySQL types
def map_dtype(dtype):
    if pd.api.types.is_integer_dtype(dtype):
        return "INT"
    elif pd.api.types.is_float_dtype(dtype):
        return "FLOAT"
    elif pd.api.types.is_datetime64_any_dtype(dtype):
        return "DATETIME"
    else:
        return "VARCHAR(255)"  # default for strings

# Step 4: Dynamically create the table based on the CSV structure
table_name = 'Orders'
create_table_query = f"CREATE TABLE {table_name} (" + ", ".join([f"{col} {map_dtype(df.dtypes[col])}" for col in df.columns]) + ")"

cursor.execute(f"DROP TABLE IF EXISTS {table_name}")  # Drop table if it already exists
cursor.execute(create_table_query)  # Create the table

# Step 5: Use SQLAlchemy engine for `df.to_sql` to insert data into MySQL
engine = create_engine('mysql+mysqlconnector://your_username:your_password@localhost/your_database')

df.to_sql(table_name, con=engine, if_exists='replace', index=False)

# Step 6: Close the connection
cnx.close()

print(f"Data from {csv_file_path} has been imported into MySQL table {table_name}")
```
